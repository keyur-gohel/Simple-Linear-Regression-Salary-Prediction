**Salary Prediction Using Years of Experiance**

*   We have created simple linear regresson to predict salary based on given years of experiance. (It's a very small dataset, and this program is for having a basic idea about linear regression) 




```python
# -*- coding: utf-8 -*-
"""
Created on Tue Sep 10 12:41:06 2019

@author: keyur
"""

# Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Getting Dataset
dataset = pd.read_csv('Salary_Data.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
# Plotting the data
dataset.plot(x = "YearsExperience", y = "Salary", kind = "scatter")
```

```python
# Spliting Training_set and and Test_set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
# Preparing Model

# Fitting The Model To Training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x, y)
```

```python
# Predict On Test_set
y_pred = regressor.predict(x_test)
```

```python
# Visulization

# Visulisation of Training set
plt.scatter(x_train, y_train, color = 'blue')
plt.plot(x_train, regressor.predict(x_train), color = 'green')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Visulisation of Test set
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'green')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
